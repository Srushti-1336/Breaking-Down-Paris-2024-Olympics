#2.Country with the highest number of athletes.

```python
#2.Country with the highest number of athletes.

country_with_most_athletes = athletes['country_long'].value_counts().head(50)
print("Top 50 Countries with the highest number of athletes:", country_with_most_athletes)

# Assuming `country_with_most_athletes` is already calculated as a Pandas Series.
# Reset the index to create a DataFrame for plotting.
top_countries = country_with_most_athletes.reset_index()
top_countries.columns = ['Country', 'Athletes']

# Set the figure size for better visualization.
plt.figure(figsize=(10, 8))

# Create a horizontal bar plot.
sns.barplot(y='Country', x='Athletes', data=top_countries, palette='viridis')

# Add titles and labels.
plt.title('Top 50 Countries with the Highest Number of Athletes', fontsize=14)
plt.xlabel('Number of Athletes', fontsize=12)
plt.ylabel('Country', fontsize=12)

# Display the plot.
plt.tight_layout()
plt.show()
```

# 3.Total Medals Won by India

```python
#3.Total Medals Won by India
#Filter the dataset for India
india_medals = medals[medals['country_long'] == 'India']

# Calculate the total number of medals
total_medals_india = india_medals['medal_type'].count()

# Calculate the number of Gold, Silver, and Bronze medals
medals_breakdown = india_medals['medal_type'].value_counts()

# Display the results
print(f"Total medals won by India: {total_medals_india}")
print("Breakdown of medals:")
print(medals_breakdown)


```

* Drop columns from Medals.csv file which are already present in Athletes.csv file.

```python
# List of columns to drop from medals.csv
columns_to_drop = ['country_code', 'country', 'country_long','gender','discipline','name']
medals = medals.drop(columns=columns_to_drop)
# Save the updated dataset if needed
medals.to_csv('updated_medals.csv', index=False)
```

* merging dataset of athletes with medals.

```python
#While merging dataset of athletes with medals on "code", In both dataset code columns were having different datatypes(Obj,int)


athletes['code'] = pd.to_numeric(athletes['code'], errors='coerce').astype('Int64')
medals['code'] = pd.to_numeric(medals['code'], errors='coerce').astype('Int64')
```

```python
# Merge athletes with medals on "code" column(Common in both dataset)
athletes_with_medals = pd.merge(athletes,medals, on="code", how="left")

print(athletes_with_medals.columns)
```

#5.Distribution of Gold, Silver, and Bronze medals among the top 10 countries.

#6.sport contributed the most medals to a single country

```python
#6.sport contributed the most medals to a single country

sport_medals = athletes_with_medals_total.groupby(['disciplines', 'country'])['medal_type'].count()
top_sport_country = sport_medals.idxmax()
print(f"Top sport-country combination: {top_sport_country}")
```

#7.Top athletes with multiple medals at different events

```python
#7.Top athletes with multiple medals at different events

# Step 1: Group by athlete name and event
top_athletes_events = (
    athletes_with_medals_total
    .groupby(['name', 'disciplines', 'event'])['medal_type']
    .count()
    .reset_index()
    .sort_values(by='medal_type', ascending=False)
)

# Step 2: Get the top athletes (by total medals)
top_athletes_names = (
    athletes_with_medals_total
    .groupby('name')['medal_type']
    .count()
    .sort_values(ascending=False)
    .head(5)
    .index
)

# Step 3: Filter the grouped data for the top athletes
top_athletes_with_events = top_athletes_events[top_athletes_events['name'].isin(top_athletes_names)]

# Step 4: Display results
print("Top athletes with multiple medals and their events/disciplines:")
print(top_athletes_with_events)
```

# 8.Calculate event distribution.

```python
# Calculate event distribution
event_distribution = athletes_with_medals_total['disciplines'].value_counts()

# Define the custom color palette
colors = ['purple', 'blue', 'green', 'orange', 'red']

# Adjust the number of colors to match the number of bars (repeating if needed)
colors = colors * (len(event_distribution) // len(colors)) + colors[:len(event_distribution) % len(colors)]

# Plot a horizontal bar chart
plt.figure(figsize=(10, 9))
event_distribution.sort_values(ascending=True).plot(
    kind='barh', color=colors[:len(event_distribution)], edgecolor="black"
)

# Add labels and title
plt.title('Event Distribution Across Sports', fontsize=16, weight='bold')
plt.xlabel('Number of Events', fontsize=14)
plt.ylabel('Events', fontsize=14)

# Add gridlines for better readability
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()

```

#11.Correlation between number of athletes and medals won

```python
# Aggregate the data
# Total number of athletes per country
athletes_count = athletes_with_medals_total.groupby('country_long')['name'].nunique().reset_index()
athletes_count.columns = ['country', 'total_athletes']

# Total medals per country
medals_count = athletes_with_medals_total.groupby('country_long')['medal_type'].count().reset_index()
medals_count.columns = ['country', 'total_medals']

# Step 2: Merge the data
athletes_medals = pd.merge(athletes_count, medals_count, on='country')

# Step 3: Correlation
correlation = athletes_medals['total_athletes'].corr(athletes_medals['total_medals'])
print(f"Correlation between number of athletes and medals won: {correlation:.2f}")

# Step 4: Visualize the data
plt.figure(figsize=(12, 8))
scatter_plot = sns.scatterplot(
    data=athletes_medals,
    x='total_athletes',
    y='total_medals',
    hue='country',
    palette='tab10',
    s=100
)
plt.title('Correlation Between Number of Athletes and Medals Won by Country', fontsize=14)
plt.xlabel('Number of Athletes', fontsize=12)
plt.ylabel('Total Medals', fontsize=12)
plt.grid(True)

# Adjust legend position to place it at the top of the plot
plt.legend(
    title="Country",
    bbox_to_anchor=(2, 1.1),  # Place above the plot
    loc='center',
    ncol=6,  # Arrange legend into 6 columns
    fontsize='small'
)


# Show the plot
plt.tight_layout()
plt.show()
```
